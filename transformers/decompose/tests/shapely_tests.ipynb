```python
%load_ext autoreload
%autoreload 2 

import torch
import itertools 
import sys

sys.path.insert(0, '..')
from decompose_util import decomp_activation_fixed, decomp_activation_all, decomp_activation
```

```python
xs = torch.randn((4, 8, 128, 768)).to(torch.torch.double)
ys_fixed = decomp_activation_fixed(xs.clone(), torch.tanh, debug=True)
ys_fixed.shape
```

```python
ys_all = decomp_activation_all(xs.clone(), torch.tanh, debug=True)
ys_all.shape
```

```python
perms = torch.tensor(list(itertools.permutations(torch.arange(4))))
ys = decomp_activation(xs.clone(), torch.tanh, debug=True, perms=perms)

assert torch.allclose(ys_all, ys)
```

```python
perms = torch.tensor(list(itertools.permutations(torch.arange(3))))
ys = decomp_activation(xs.clone(), torch.tanh, debug=True, perms=perms)

assert torch.allclose(ys_fixed, ys)
```
