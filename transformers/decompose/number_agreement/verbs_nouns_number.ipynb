```python
import pandas as pd
```

```python
path1 = "../../../datasets/gwordlist/original/frequency-alpha-alldicts.txt"
path2 = "../../../datasets/gwordlist/frequency-all.txt"

all_words = list(map(
            lambda x: x.split(),
            open(path1).readlines()
        ))[1:]

words_dict = {}
for _, word, count, _, _ in all_words:
    count = int(count.replace(",", ""))
    # filter out words that are never used
    if count > 1000:
        words_dict[word] = count

len(words_dict)
```

```python
words_bypos = {}
# https://books.google.com/ngrams/info
parts_of_speech = ["NOUN", "VERB", "ADJ", "ADV", "PRON", "DET", "ADP", "NUM", "CONJ", "PRT"]

with open(path2) as f:
    for line in f:
        if line == "":
            continue

        _, w, count, _, _ = line.split()
        # skip header and non-words
        if w[0] == "_" or count == "COUNT":
            continue
        
        count = int(count.replace(",", ""))
        try:
            word, pos = w.split("_")
        except Exception as e:
            continue

        if word not in words_dict or pos not in parts_of_speech:
            continue
        
        if word not in words_bypos:
            words_bypos[word] = {
                "_TOTAL_": words_dict[word],
                pos: count / words_dict[word]
            }
        else:
            words_bypos[word][pos] = count / words_dict[word]

words_bypos
```

```python
words_df = pd.DataFrame.from_records(words_bypos).T.fillna(0).sort_values(by="_TOTAL_", ascending=False)
words_df.index.name = "WORD"
words_df.to_csv("all_words_by_pos.csv")

words_df
```

```python
for pos in parts_of_speech:
    # form has to make up at least 20% of appearences
    pos_df = words_df.loc[words_df[pos] > 0.8][["_TOTAL_", pos]]
    print(pos, len(pos_df))
    pos_df.to_csv(f"all_{pos}s.csv")
```
